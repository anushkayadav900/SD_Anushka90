```python
def arithmetic(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b

print(arithmetic(10, 5, '+'))
print(arithmetic(10, 5, '-'))
print(arithmetic(10, 5, '*'))
print(arithmetic(10, 5, '/'))
```

***GENERATIVE***   ***FUNCTION***

```python
def test_fib(n):
  a,b=0,1
  for i in range(n):
    yield a
    a,b=b,a+b
```

```python
for i in test_fib(7):
  print(i)
```

```python
def test_fib1():
  a,b=0,1
  while True:
    yield a
    a,b=b,a+b
```

```python
fib=test_fib1()
```

```python
print(fib)
```

```python
for i in range(10):
  print(next(fib))
```
