```python
def count_test(n):
  count=1
  while count<=n:
    yield count
    count= count + 1
```

```python
c = count_test(5)
```

```python
print(c)
```

```python
type(c)
```

```python
for i in c:
  print(i)
```

```python
total = 0

n = int(input("Enter number of expenses: "))

for i in range(n):
    expense = float(input("Enter expense: "))
    total += expense

print("Total Expense =", total)
```

```python
expenses = {}
total = 0

n = int(input("Enter number of expenses: "))

for i in range(n):
    name = input("Expense name: ")
    amount = float(input("Amount: "))
    expenses[name] = amount
    total += amount

print("\nExpenses:")
for name, amount in expenses.items():
    print(name, "=", amount)

print("\nTotal Expense =", total)
```
